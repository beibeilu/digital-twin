```python
tract = geopandas.read_file('Source/2010 Census Tracts.geojson')
tract.head()
```

```python
print(type(tract["geometry"]))
```

```python
tract['GeoId'] = tract['puma'] + tract['boro_ct2010']


convert_dict = {
                'puma': int,
                'ct2010': int,
                'boro_code':int,
                'boro_ct2010':int,
                'GeoId':int,
               }
df = tract.astype(convert_dict)
print(df.dtypes)
```
